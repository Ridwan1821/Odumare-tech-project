```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = pd.ExcelFile('skills.xlsx')
df = file.parse('skills')
df.head()
```

```python
df.info()
```

```python
df1 = pd.DataFrame(df['Course of Study'].value_counts())
df1 = df1.reset_index().rename(columns={'index':'course_of_study', 'Course of Study': 'count'})
df1 = df1.sort_values(by = 'count')
plt.figure(figsize=(8,10))
plt.barh(df1['course_of_study'], df1['count'])
plt.xticks(rotation=90);
```
